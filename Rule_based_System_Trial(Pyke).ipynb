```python
# Activate the rule-based system

import contextlib
import sys

engine = knowledge_engine.engine('/content/drive/MyDrive/MSBA 316 NLP/Project')

def bc_test_questions():
    engine.reset()
    engine.activate('bc_simple_rules_questions')
    try:
        with engine.prove_goal('bc_simple_rules_questions.recommend($ans)') as gen:
            for vars, plan in gen:
                print("You should study: %s" % (vars['ans']))
    except Exception:
        krb_traceback.print_exc()
        sys.exit(1)
```

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig, AutoTokenizer
from peft import PeftModel, PeftConfig

base_model = "unsloth/llama-3-8b-bnb-4bit"
adapter_model = "/content/drive/MyDrive/MSBA 316 NLP/Project/CoCoT_tuned" # Specify the path to your saved fine-tuned model


bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
)

model = AutoModelForCausalLM.from_pretrained(
    base_model,
    quantization_config=bnb_config,
    trust_remote_code=True,
    torch_dtype=torch.bfloat16
)

model = PeftModel.from_pretrained(model, adapter_model, quantization_config=bnb_config, torch_dtype=torch.bfloat16, device_map="auto")

tokenizer = AutoTokenizer.from_pretrained(adapter_model)

model.config.use_cache = False
```

```python
# Initialize a text generation pipeline with the specified model and tokenizer

import transformers
import torch

text_generation_pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    device_map="auto",
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    max_new_tokens=500,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
)
```

```python
# Create a LangChain pipeline using the created Hugging Face text generation pipeline

from langchain import HuggingFacePipeline

llm = HuggingFacePipeline(pipeline=text_generation_pipeline)
```

```python
from langchain.agents import initialize_agent, Tool
from langchain.agents import AgentType
from langchain.tools import tool
```

```python
#Define a custom tool that activates the rule-based system

@tool
def process_query(query):
  """Uses the rule-based system"""
  inference_result = bc_test_questions()
  return inference_result
```

```python
class PykeInferenceTool:
    def __init__(self, engine):
        self.engine = engine

    def __call__(self, query):
        return process_query(query)
```

```python
pyke_tool = PykeInferenceTool(engine)
```

```python
# Define the tools that include both the rule-based system and the language model

from langchain.agents import initialize_agent, Tool
from langchain.agents import AgentType
from langchain.tools import BaseTool, StructuredTool, tool

tools = [
    StructuredTool.from_function(
        name="CoCoT",
        func=llm.invoke,
        description="Use this for general Python tutoring questions"
    ),
    StructuredTool.from_function(
        name="PykeInference",
        func=pyke_tool.__call__,
        description="Use this for detailed code analysis and rule-based explanations"
    )
]
```

```python
from langchain.agents import initialize_agent, Tool, AgentType
from langchain.prompts import PromptTemplate


agent_prompt = """You are a Python tutor with access to two tools:
1) A language model trained for Python tutoring: CoCoT
2) A rule-based inference system for detailed code analysis

For each student question, decide which tool(s) to use. You may use both if necessary.
Always provide a final, integrated response that combines insights from the tools you used.

Student question: {input}

Think step-by-step:
1) Which tool(s) should I use for this question?
2) What information do I need from each tool?
3) How can I combine the information into a coherent response?

Begin your analysis:

{agent_scratchpad}
"""

# Create a prompt template
prompt = PromptTemplate(
    template=agent_prompt,
    input_variables=["input", "agent_scratchpad"]
)

# Initialize the agent with the prompt
myagent = initialize_agent(
    tools,
    llm,
    verbose=False,
    handle_parsing_errors=True,
    agent_kwargs={"prompt": prompt}
)

# Call the agent with a certain prompt
def integrated_tutor(question):
    try:
        response = myagent.invoke(input=question)
        return response
    except Exception as e:
        return f"An error occurred: {str(e)}"
```

```python
# Example usage

question = "Can you explain how to use list comprehensions and analyze this code: [x**2 for x in range(10) if x % 2 == 0]?"
result = integrated_tutor(question)
print(result)
```
